```python
url = "https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/mobile_dataset.csv"
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.info()
```

```python
### check for duplicates

duplicated = df.duplicated().value_counts()
duplicated
```

```python
#### Univariate Selection

X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
df.price_range.value_counts()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df.shape
```

```python
### Apply SelectKBest Algorithm

orderd_rank_features = SelectKBest(score_func=chi2, k=20)
ordered_feature = orderd_rank_features.fit(X, y)
```

```python
ordered_feature
```

```python
ordered_feature.scores_
```

```python
dfscores = pd.DataFrame(ordered_feature.scores_, columns=["Score"])
dfcolumns = pd.DataFrame(X.columns)
```

```python
features_rank = pd.concat([dfcolumns, dfscores], axis=1)
```

```python
features_rank
```

```python
features_rank.columns=["Features", "Score"]
```

```python
features_rank
```

```python
features_rank.nlargest(10, 'Score')
```

#### Feature Importance
This technique gives a score for each feature of data, higher the score more relevant it is. 

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
```

```python
model = ExtraTreesClassifier()
model.fit(X, y)
```

```python
print(model.feature_importances_)
```

```python
ranked_features = pd.Series(model.feature_importances_, index=X.columns)
ranked_features.nlargest(10).plot(kind='bar')
plt.show()
```

```python
### Remove the correlated features 

threshold = 0.5
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df.iloc[:, :-1], threshold)
```
